# Acceleration data

```python
import json
import math

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import Scatter, Layout
init_notebook_mode(connected=True)
```

```python
accelData=[]
with open('coding-agents.raw', 'r') as accelDataFile:
    accelData=accelDataFile.readlines()
```

```python
x=[json.loads(a.split('\'')[1])['x'] for a in accelData]
y=[json.loads(a.split('\'')[1])['y'] for a in accelData]
z=[json.loads(a.split('\'')[1])['z'] for a in accelData]
```

```python
total={}
for index in range(len(x)):
    total2=[]
    for index2 in range(len(x[index])):
        total2.append(math.sqrt(x[index][index2]**2+y[index][index2]**2+z[index][index2]**2))
    totalsum=sum(total2)
    if totalsum<10*60:
        total['{}_lazy'.format(index)]=total2
    else:
        if totalsum<10.5*60:
            total['{}_normal'.format(index)]=total2
        else:
            total['{}_active'.format(index)]=total2
```

```python
data=[]
for t in total.keys():
    data.append(Scatter(y=total[t], name=t))

iplot({
    'data': data,
    'layout': Layout(
        title='Acceleration profile of selected app users',
        yaxis=dict(title='Acceleration [m/s^2]',),
        xaxis=dict(title='Time since notification [s]',)
    )
})
```
